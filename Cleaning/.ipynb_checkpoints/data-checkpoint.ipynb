```python
import pandas as pd
import re
```

```python
df = pd.read_csv('../Data/data_cleaned_2021.csv')
df.head()
```

```python
df_temp = df[['Salary Estimate', 'Rating', 'Size', 'Job Title']].copy()
df_temp.head()
```

```python
df_temp['Salary_string'] = df_temp['Salary Estimate'].apply(lambda x: x.split(' ')[0])
df_temp['Salary_list'] = df_temp['Salary_string'].apply(lambda x: x.split('-'))

def sort_min(x):
    if len(x) != 2:
        return None
    else:
        return int(re.findall(r'\d+', x[0])[0])
df_temp['Salary_min'] = df_temp['Salary_list'].apply(sort_min)

def sort_max(x):
    if len(x) != 2:
        return None
    else:
        return int(re.findall(r'\d+', x[1])[0])
df_temp['Salary_max'] = df_temp['Salary_list'].apply(sort_max)

df_temp['Salary_mean'] = df_temp.apply(lambda x: (x[6] + x[7])/2, axis = 1)

df_temp.head()
```
